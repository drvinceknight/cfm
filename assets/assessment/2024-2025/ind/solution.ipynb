b. Consider the initial sizes of the armies: $x(0) = 1000$ and $y(0) = 500$ and the fact that $\alpha=-4/5$ and $\beta=-11/10$. Output the particular solutions (in the same format as for question a.):

_Available marks: 12_

````python
### BEGIN SOLUTION
conditions = {x(0): 1000, y(0): 500}

equations = [
    sym.Eq(sym.diff(x(t), t), - 4 / sym.S(5) * y(t)),
    sym.Eq(sym.diff(y(t), t), - 11 / sym.S(10) * x(t)),
]
X, Y = sym.dsolve(equations, set=True, ics=conditions)
X, Y
### END SOLUTION
````
