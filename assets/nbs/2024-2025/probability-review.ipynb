## Coursework like question

We are going to solve the coursework like problem from https://vknight.org/cfm/assets/handouts/autumn/05-probability/main.pdf

The first question asks us to create a function to reproduce the experiment. One way to do this is below:

```python
import random

def sample_experiment():
    """
    This returns 2 variables:
        - The bowl type (as a string, either "spin" or "pace")
        - Whether or not the ball was hit (as a boolean)
    """
    if random.random() < .45:
        bowl = "spin"
        is_hit = random.random() < .4
    else:
        bowl = "pace"
        is_hit = random.random() < .35
    return bowl, is_hit
```

A slightly cleaner/clearer/preffered is:

```python
import random

def sample_experiment():
    """
    This returns 2 variables:
        - The bowl type (as a string, either "spin" or "pace")
        - Whether or not the ball was hit (as a boolean)
    """
    if random.random() < .45:
        bowl = "spin"
        is_hit = random.random() < .4
        return bowl, is_hit
    bowl = "pace"
    is_hit = random.random() < .35
    return bowl, is_hit
```

The second question asks us to use 1,000,000 samples to approximate the probability of a ball being hit:

```python
number_of_repetitions = 1_000_000
samples = [sample_experiment() for _ in range(number_of_repetitions)]
samples
```

```python
probability_of_being_hit = sum(is_hit for (bowl, is_hit) in samples) / number_of_repetitions
probability_of_being_hit
```

We can confirm this by calculating it:

$$
P(\text{H}) = P(\text{H}|\text{Sp})\times P(\text{Sp}) + P(\text{H}|\text{P})\times P(\text{P})
$$

Where the events are:

- $\text{H}$: A hit;
- $\text{Sp}$: The ball is bowled with spint;
- $\text{P}$: The ball is bowled with pace.

This corresponds to:

$$
P(\text{H}) = .4\times .45 + .35\times .55
$$

```python
.4 * .45 + .35 * .55
```

The final question asks us to approximate the probability of the ball being bolwed with spin given that it was hit. First let us get the samples that correspond to a hit:

```python
balls_with_hit = [ball for (ball, is_hit) in samples if is_hit is True]
balls_with_hit
```

Now let us count them:

```python
number_of_hits = len(balls_with_hit)
number_of_hits
```

Now we can count the balls that were bowled with spin.

```python
probability_of_spin_given_hit = sum(ball == "spin" for ball in balls_with_hit) / number_of_hits
probability_of_spin_given_hit
```

We can now confirm this using Bayes' theorem:

$$
P(\text{Sp}|\text{H})P(\text{H}) = P(\text{H}|\text{Sp})P(\text{Sp})
$$

which gives:

$$
P(\text{Sp}|\text{H}) = \frac{P(\text{H}|\text{Sp})P(\text{Sp})}{P(\text{H})}
$$

```python
.4 * .45 / 0.3725
```
