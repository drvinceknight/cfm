## Variables, conditionals and loops

Example:

A perfect number is a positive integer whose divisors (excluding itself) sum to
itself. For example, $6$ is a perfect number because $6 = 1 + 2 +
3$.

1. How many perfect numbers are there less than 20?
2. What is the next perfect number?

```python
def is_perfect(n):
    """
    Checks if a given integer n is perfect or not.

    It returns True if perfect, False if not.

    A number is perfect if the sum of it's divisors is equal to the number itself.

    For example 6 = 3 + 2 + 1
    """
    return sum(i for i in range(1, n) if (n % i == 0)) == n
```

```python
is_perfect(n=6)
```

```python
is_perfect(n=5)
```

```python
checks = [is_perfect(n=value) for value in range(1, 20)]
sum(checks)
```

```python
n = 496 + 1
while (is_perfect(n) is False):
    n = n + 1
```
